```python
import sys
import numpy as np
from PyQt5.QtWidgets import QApplication, QWidget, QTableWidget, QTableWidgetItem, QVBoxLayout
from PyQt5.QtCore import QTimer

class MyApp(QWidget):
    def __init__(self, data_array):
        super().__init__()
        self.data_array = data_array
        self.current_row_index = 0
        self.initUI()

    def initUI(self):
        self.tableWidget = QTableWidget()
        self.tableWidget.setRowCount(20)  # Number of elements in each row of data_array
        self.tableWidget.setColumnCount(1)  # Single column

        # Timer setup
        self.timer = QTimer(self)
        self.timer.timeout.connect(self.updateTable)
        self.timer.start(1000)  # Update interval in milliseconds

        # Layout setup
        layout = QVBoxLayout()
        layout.addWidget(self.tableWidget)
        self.setLayout(layout)

        self.setWindowTitle('Data Display')
        self.setGeometry(300, 300, 200, 400)
        self.show()

    def updateTable(self):
        if self.current_row_index < len(self.data_array):
            # Update the table with the current row of data_array
            for i, value in enumerate(self.data_array[self.current_row_index]):
                self.tableWidget.setItem(i, 0, QTableWidgetItem(str(value)))

            self.current_row_index += 1
        else:
            self.timer.stop()  # Stop the timer when all data is displayed

if __name__ == '__main__':
    app = QApplication(sys.argv)

    # Example data array of shape (n, 20)
    n = 10  # Number of rows
    data_array = np.random.randint(0, 100, (n, 20))

    ex = MyApp(data_array)
    sys.exit(app.exec_())
```
